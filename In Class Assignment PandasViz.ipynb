```python
import pandas as pd
from pandas_datareader import data, wb
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

Here I am opening the Amazon analytics of January 1st to today.

```python
start1 = datetime.datetime(2022, 1, 1)
end1 = datetime.datetime(2022, 2, 15)
```

```python
AMZN1 = data.DataReader("AMZN", 'yahoo', start1, end1)
```

```python
myAMZN1 = AMZN1.reset_index()
```

# Part Ai & Aii: 
## Here I am using ggplot and building a time series.

```python
plt.style.use('ggplot')
myAMZN1.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

# Part Aiii: Here I am building a new timeline of data to create a bar graph.

```python
start2 = datetime.datetime(2021, 12, 1)
end2 = datetime.datetime(2021, 12, 31)
AMZN2 = data.DataReader("AMZN", 'yahoo', start2, end2)
myAMZN2 = AMZN2.reset_index()
```

### Here is my code to create the bar graph.

```python
myAMZN2.plot.bar(x='Date',y='Open',figsize=(12,3),lw=1)
```
